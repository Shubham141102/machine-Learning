```python
df = df.rename(columns={'town ': 'town'})
```

```python
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
# joining the dataframe with the help of concat function
merged = pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final = merged.drop(['town','west windsor'],axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
x = final.drop('price',axis='columns')
```

```python
x
```

```python
y = final.price
```

```python
y
```

```python
reg.fit(x,y)
```

```python
reg.predict([[2000,0,1]])
# this will the price of the house of 2000 sqrfeet in robbinsville
```

```python
# to understand how perfect the model is use the foloowing function
reg.score(x,y)
# this states that the model is 95% accurate
```
